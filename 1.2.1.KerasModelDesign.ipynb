````python
# 변수의 shape, type, value 확인
def p(var,_name="") :
    if _name != "" : print(f'<<{_name}>>')
    if type(var)!=type([]):
        try:
            print(f'Shape:{var.shape}')
        except :
            pass
    print(f'Type: {type(var)}')
    print(f'Values: {var}')

def pst(_x,_name=""):
    print(f'<<{_name}>> Shape{_x.shape}, {type(_x)}')
def ps(_x,_name=""):
    print(f'<<{_name}>> Shape{_x.shape}')
````

#### 2.2. Functional API

````python
keras.backend.clear_session()

# 1. DataFlow(Graph)정의
input_x = keras.Input(shape=(28,28))
x0 = layers.Flatten()(input_x)
x1 = layers.Dense(32, activation="relu")(x0)
x2 = layers.Dense(32, activation="relu")(x0)
x3 = layers.concatenate([x1, x2])
output_x = layers.Dense(10,activation="softmax")(x3)

# 2. 모델 인스턴스 : input,output으로 DataFlow를 추적
model = keras.Model(inputs=input_x, outputs=output_x)

model.summary()
keras.utils.plot_model(model,show_shapes=True,show_layer_activations=True,dpi=70)
````

````python
# train option 설정
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
# training
history = model.fit(x_train, y_train, epochs=3, batch_size=128)
````

#### 2.3. SubClassing API

````python
@keras.utils.register_keras_serializable() ## save에서 필요
class MyModel(keras.Model):
    def __init__(self, *args, **kwargs):   ## save에서 *args, **kwargs 필요
        super(MyModel, self).__init__(*args, **kwargs) ##
        self.flatten = layers.Flatten()
        self.d1 = layers.Dense(32, activation="relu")
        self.d2 = layers.Dense(32, activation="relu")
        self.d3 = layers.Dense(10, activation="softmax")

    def call(self, x):
        x0 = self.flatten(x)
        x1 = self.d1(x0)
        x2 = self.d2(x0)
        c1 = layers.concatenate([x1, x2])
        return self.d3(c1)
````

- 모델 인스턴스 생성후 사용

````python
keras.backend.clear_session()

# 1. model instance
model = MyModel()
# 2. DataFlow(Graph)정의
input_x = keras.Input(shape=(28,28))
output_x = model(input_x)

model.summary()
## .utils.plot_model() : Sequntial/Functional API를 위한 것임
keras.utils.plot_model(model,show_shapes=True,dpi=70)
````

**build method로 input shape 변경하기**

````python
# 사용자 정의 모델 클래스
class CustomModel(keras.Model):
    def __init__(self):
        super(CustomModel, self).__init__()
        self.dense1 = layers.Dense(64, activation='relu')
        self.dense2 = layers.Dense(10, activation='softmax')
    def call(self, inputs):
        x = self.dense1(inputs)
        return self.dense2(x)
    def build(self, input_shape):
        # 입력 형태를 기반으로 모델의 파라미터를 초기화
        self.dense1.build(input_shape)
        self.dense2.build((None, 64))  # 첫 번째 Dense 레이어의 출력 형태

model = CustomModel()
model.build((None, 32))
model.summary()
````
